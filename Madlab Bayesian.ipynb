```python
def p_s(N_t,b_t,N_j,b_i):
    i = int((b_i)/ad_step)
    j = int((N_j-ssmin)/ss_step)
    return Z[i][j]
```

```python
def p_u(N_t,b_t,N_j,b_i):
    return 1-p_s(N_t,b_t,N_j,b_i)
```

```python
def p_s_p(N_t,b_t,N_j,b_i):
    num= p_p_s(N_t,b_t,N_j,b_i)*p_s(N_t,b_t,N_j,b_i)
    den= p_p_s(N_t,b_t,N_j,b_i)*p_s(N_t,b_t,N_j,b_i)+p_p_u(N_t,b_t,N_j,b_i)*p_u(N_t,b_t,N_j,b_i)
    return num/den
```

```python
def p_s_m(N_t,b_t,N_j,b_i):
    num= p_m_s(N_t,b_t,N_j,b_i)*p_s(N_t,b_t,N_j,b_i)
    den= p_m_s(N_t,b_t,N_j,b_i)*p_s(N_t,b_t,N_j,b_i)+p_m_u(N_t,b_t,N_j,b_i)*p_u(N_t,b_t,N_j,b_i)
    return num/den
#     return 1-p_s_p(N_t,b_t,N_j,b_i)
```

```python
def update_grid_stable(N_t,b_t):
    t_1 = int((b_t)/ad_step)
    t_2 = int((N_t-ssmin)/ss_step)
    for i in range(row):
        for j in range(col):
            if i==t_1 and j==t_2:
                continue
            N_j= j*ss_step+(ssmin)
            b_i= i*ad_step
            Z[i][j]=p_s_p(N_t,b_t,N_j,b_i)
    
```

```python
def update_grid_unstable(N_t,b_t):
    t_1 = int((b_t)/ad_step)
    t_2 = int((N_t-ssmin)/ss_step)
    for i in range(row):
        for j in range(col):
            if i==t_1 and j==t_2:
                continue
            N_j= j*ss_step+(ssmin)
            b_i= i*ad_step
            Z[i][j]=p_s_m(N_t,b_t,N_j,b_i)
```
